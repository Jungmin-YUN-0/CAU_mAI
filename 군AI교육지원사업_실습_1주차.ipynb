## 2-2. Subword Tokenizer

### **# BPE(Byte Pair Encoding)**

* 1994년에 제안된 데이터 압축 알고리즘

* 이후 Subword Segmentation Algorithm으로 응용

* 기본적으로, 연속적으로 가장 많이 등장한 글자 쌍을 찾아 하나의 글자로 병합하는 방식 수행



**[예제]**

**aaabdaaabac**


1)

가장 자주 등장하는 byte pair는 'aa' 

→'aa'라는 byte pair를 하나의 byte인 'Z'로 치환

**ZabdZabac** (Z=aa)




2)

가장 자주 등장하는 byte pair는 'ab' ('Za'도 있지만, 치환되지 않은 byte에 우선순위를 부여하는 것으로 가정)

→'ab'라는 byte pair를 하나의 byte인 'Y'로 치환

**ZYdZYac** (Z=aa, Y=ab)




3)

가장 자주 등장하는 byte pair는 'ZY' 

→'ZY'라는 byte pair를 하나의 byte인 'X'로 치환

**XdXac** (Z=aa, Y=ab, X=ZY)




4)

더 이상 병합할 byte pair가 없으므로, 위의 결과를 최종 결과로 하여 BPE 종료

### **# 자연어 처리에서의 BPE(Byte Pair Encoding)**

* 자연어 처리에서의 BPE는 Subword Segmentation Algorithm임.

* 글자(character) 단위에서 점차적으로 단어 집합(vocabulary)를 만들어 내는 Bottom up 방식의 접근을 사용

**[예제]**

low, lower, newest, widest

**1) 기존의 접근(빈도 기반의 카운트)**

```python
'''
# dictioanry (훈련 데이터에 있는 단어와 등장 빈도수)
  → low : 5, lower : 2, newest : 6, widest : 3

# vocabulary
  → low, lower, newest, widest
'''
```

```python
# 이때 테스트 과정에서 'lowest'라는 단어가 등장한다면?
#'lowest'를 학습한 적이 없으므로, 이에 대응하지 못하는 OOV 문제 발생 !
```

**2) BPE 알고리즘**

```python
'''
# dictionary (모든 단어들을 character 단위로 분리)
  → l o w : 5,  l o w e r : 2,  n e w e s t : 6,  w i d e s t : 3

# vocabulary (초기 구성: 글자 단위로 분리된 상태)
  → l, o, w, e, r, n, w, s, t, i, d
'''
```

BPE는 사용자가 정한 단어 집합의 크기를 만족할 때까지 알고리즘의 동작을 반복함.
(가장 빈도수가 높은 유니그램의 쌍을 하나의 유니그램으로 통합하는 과정을 반복)

1회 - 딕셔너리를 참고하여, 빈도수가 9로 가장 높은 (e, s)의 쌍을 es로 통합

```python
# (e, s), (s, t) -> 둘다 빈도수가 9
# 실제 BPE에서는 병합 우선순위가 존재.
# 여기서는 먼저 나온 pair에 우선순위 부여하는 것으로 설정
```

```python
'''
# dictionary update!
l o w : 5,
l o w e r : 2,
n e w es t : 6,
w i d es t : 3

# vocabulary update!
l, o, w, e, r, n, w, s, t, i, d, es
'''
```

2회 - 빈도수가 9로 가장 높은 (es, t)의 쌍을 est로 통합

```python
'''
# dictionary update!
l o w : 5,
l o w e r : 2,
n e w est : 6,
w i d est : 3

# vocabulary update!
l, o, w, e, r, n, w, s, t, i, d, es, est
'''
```

3회 - 빈도수가 7로 가장 높은 (l, o)의 쌍을 lo로 통합

```python
'''
# dictionary update!
lo w : 5,
lo w e r : 2,
n e w est : 6,
w i d est : 3

# vocabulary update!
l, o, w, e, r, n, w, s, t, i, d, es, est, lo
'''
```

10회 반복했을 때

```python
'''
# dictionary update!
low : 5,
low e r : 2,
newest : 6,
widest : 3

# vocabulary update!
l, o, w, e, r, n, w, s, t, i, d, es, est, lo, low, ne, new, newest, wi, wid, widest
'''
```

```python
# 이때 테스트 과정에서 'lowest'라는 단어가 등장한다면?
# 기존 방식) OOV 문제 발생
# BPE)
#  lowest를 글자 단위로 분해 → l, o, w, e, s, t
#  업데이트된 단어집합을 통해, low와 est를 찾아냄
#  즉 lowest는 OOV가 아니라, low와 est 두 단어로 인코딩됨
```

**코드실습**

```python
dictionary = {'l o w' : 5,
         'l o w e r' : 2,
         'n e w e s t':6,
         'w i d e s t':3
         }
```

```python
import re, collections

def get_stats(dictionary):
    # 유니그램의 pair들의 빈도수를 카운트
    pairs = collections.defaultdict(int)
    for word, freq in dictionary.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i],symbols[i+1]] += freq
    print('현재 pair들의 빈도수 :', pairs)
    return pairs


def merge_dictionary(pair, v_in):    # 'n e w e s t' → 'n e w es t'
    v_out = {}
    bigram = re.escape(' '.join(pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        v_out[w_out] = v_in[word]
    return v_out
```

```python
from IPython.display import display, Markdown, Latex

num_merges = 10    # BPE 수행 횟수

bpe_codes = {}
bpe_codes_reverse = {}

for i in range(num_merges):
    display(Markdown("### Iteration {}".format(i + 1)))
    pairs = get_stats(dictionary)
    best = max(pairs, key=pairs.get)
    dictionary = merge_dictionary(best, dictionary)

    bpe_codes[best] = i
    bpe_codes_reverse[best[0] + best[1]] = best

    print("new merge: {}".format(best))
    print("dictionary: {}".format(dictionary))
```

OOV 처리


[low, lower, newest, widest] (+ lowest)

```python
def get_pairs(word):
    """ex_ 1st iteration
    (l,o,w,e,s,t) → {('e', 's'), ('l', 'o'), ('o', 'w'), ('s', 't'), ('w', 'e')}
    """
    pairs = set()
    prev_char = word[0]
    for char in word[1:]:
        pairs.add((prev_char, char))
        prev_char = char
    return pairs
```

```python
def encode(orig):
    """Encode word based on list of BPE merge operations, which are applied consecutively"""

    word = tuple(orig)    #lowest → (l,o,w,e,s,t)
    display(Markdown("__word split into characters:__ <tt>{}</tt>".format(word)))

    pairs = get_pairs(word)    # bigrams in the word

    iteration = 0
    while True:
        iteration += 1
        display(Markdown("__Iteration {}:__".format(iteration)))

        print("bigrams in the word: {}".format(pairs))
        bigram = min(pairs, key = lambda pair: bpe_codes.get(pair, float('inf')))    #bpe_codes.get(pair, float('inf)): pair에 대한 키 반환. pair가 없으면 float('inf')를 반환.
        print("candidate for merging: {}".format(bigram))
        if bigram not in bpe_codes:    # False인 경우에 동작.
            display(Markdown("__Candidate not in BPE merges, algorithm stops.__"))
            break
        first, second = bigram    # (e, s)
        new_word = []
        i = 0
        while i < len(word):
            try:
                j = word.index(first, i)    # i번째 위치부터 first가 위치한 자리
                new_word.extend(word[i:j])
                i = j
            except:
                new_word.extend(word[i:])
                break

            if word[i] == first and i < len(word)-1 and word[i+1] == second:
                new_word.append(first+second)
                i += 2
            else:
                new_word.append(word[i])
                i += 1
        new_word = tuple(new_word)
        word = new_word

        print("word after merging: {}".format(word))
        if len(word) == 1:
            break
        else:
            pairs = get_pairs(word)



    return word
```

**Dictionary 자료형 사용**

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a hug secret to himself was dribing the barber crazy. the barber went up a huge mountain."
```

```python
# sentence tokenization

sentences = sent_tokenize(raw_text)
print(sentences)
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')[:20]
```

```python
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
  # word tokenization
  tokenized_sentence = word_tokenize(sentence)
  result = []
  for word in tokenized_sentence:
    word = word.lower()    # 소문자화
    if word not in stop_words:    # 토큰화된 단어에 대하여, 불용어 제거
      if len(word) > 2:    # 길이가 2 미만인 단어 제거
        result.append(word)
        if word not in vocab: 
          vocab[word] = 0
        vocab[word] += 1
  preprocessed_sentences.append(result)
print(preprocessed_sentences)
```

```python
# vocab <- 각 단어에 대한 빈도수 저장
print('vocabulary: ', vocab)
```

```python
print(vocab['barber'])    # 단어 'barber'의 빈도수 출력
```

```python
# 빈도수가 높은 순서대로 정렬

vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse = True)
print(vocab_sorted)
```

```python
# 높은 빈도수를 가진 순서대로 단어에 정수 부여(1부터)

word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted:
  if frequency > 1:    # 빈도수 1 이하의 단어는 제외
    i = i+1
    word_to_index[word] = i

print(word_to_index)
```

```python
# 빈도수 기준, 상위 5개의 단어만 남기기

vocab_size = 5

words_frequency = [word for word, index in word_to_index.items() if index >= vocab_size + 1]    # 인덱스가 5 초과인 단어 제거

for w in words_frequency:    # 해당 단어에 대한 인덱스 정보 삭제
  del word_to_index[w]

print(word_to_index)
```

```python
# 문장 내 단어를 정수로 바꾸기 

preprocessed_sentences    # 토큰화가 이루어진 sentences
```

**OOV(Out-Of-Vocabulary) 문제**
_ 단어 집합에 존재하지 않는 단어가 발생

```python
# word_to_index에 'OOV'를 추가해서, 단어 집합에 없는 단어들을 'OOV'의 인덱스로 인코딩

word_to_index['OOV'] = len(word_to_index) + 1
print(word_to_index)
```

```python
# word_to_index를 사용하여, sentences의 모든 단어들을 맵핑되는 정수로 인코딩

encoded_sentences = []
for sentence in preprocessed_sentences:
  encoded_sentence = []
  for word in sentence:
    try:
      encoded_sentence.append(word_to_index[word])    # 단어 집합에 존재하는 단어라면, 해당 단어의 정수를 리턴.
    except KeyError:
      encoded_sentence.append(word_to_index['OOV'])    # 단어 집합에 존재하지 않는 단어라면, 'OOV'의 정수를 리턴.
  encoded_sentences.append(encoded_sentence)

print(encoded_sentences)
```

* 단어를 One-Hot Vector로 표현
* 이때 벡터의 차원은 단어 집합(vocabulary: 서로 다른 단어들의 집합)의 크기

**[One-Hot Encoding Process]**

1. 각 단어에 고유한 인덱스를 부여(정수 인코딩)
2. 대상 단어의 인덱스 위치에 1을 부여하고, 다른 단어의 인덱스 위치에는 0을 부여

```python
data = ['I love apple','I love banana', 'I hate apple','I hate banana']
token_index = {}

for sample in data:
    for word in sample.split():    # 띄어쓰기 기준으로 토큰화
        if word not in token_index:
            token_index[word] = len(token_index) + 1     # 인덱스는 1부터 시작

token_index
```

```python
import torch

max_length = 3    # 한 문장에 단어가 3개씩 존재

results = torch.zeros([len(data), max_length, max(token_index.values()) + 1])

# print(token_index.values())    # dict_values([1, 2, 3, 4, 5])
print(results)    # 결과를 저장할 배열
```

```python
for i, sample in enumerate(data):    # (idx, sentence) 
    for j, word in list(enumerate(sample.split())):    # 1st loop: [(0, 'I'), (1, 'love'), (2, 'apple')] 
        index = token_index.get(word)     
        results[i, j, index] = 1    #i(문장 인덱스), j(문장 내 단어 인덱스), index(token_index의 value값)

print(results)
```

## 3-2. Bag of Words

**[Bag of Words]**

* (Count-based word representation)
* 단어의 출현 빈도(frequency)에 기반한 텍스트 데이터 수치화 표현법
* 단어의 순서는 고려하지 않음

**[Bag of Words Process]**



1.   각 단어에 고유한 정수 인덱스 부여 (단어 집합 생성)
2.   각 인덱스에 위치한 단어 토큰 등장 횟수를 기록한 벡터 생성

* **[문서 2]** John also likes to watch football games.

```python
stopwords.words('english')

def build_bag_of_words(document):
  document = document.replace('.', '')    # 온점 제거
  tokenized_document = word_tokenize(document)

  word_to_index = {}
  bow = []

  for word in tokenized_document:  
    if word not in word_to_index.keys():
      word_to_index[word] = len(word_to_index)  
      # BoW에 전부 기본값 1을 넣는다.
      bow.insert(len(word_to_index) - 1, 1)
    else:
      # 재등장하는 단어의 인덱스
      index = word_to_index.get(word)
      # 재등장한 단어는 해당하는 인덱스의 위치에 1을 더한다.
      bow[index] = bow[index] + 1

  return word_to_index, bow
```

```python
vocab, bow = build_bag_of_words(doc3)
print('vocabulary :', vocab)
print('bag of words vector :', bow)
```

* BoW는 각 단어가 등장한 횟수를 수치화하는 텍스트 표현 방법이므로 주로 어떤 단어가 얼마나 등장했는지를 기준으로 문서가 어떤 성격의 문서인지를 판단하는 작업에 많이 사용됨.

* 즉, 분류 문제나 여러 문서 간의 유사도를 구하는 문제에 주로 쓰임. 가령, '달리기', '체력', '근력'과 같은 단어가 자주 등장하면 해당 문서를 체육 관련 문서로 분류할 수 있을 것이며, '미분', '방정식', '부등식'과 같은 단어가 자주 등장한다면 수학 관련 문서로 분류할 수 있음.

**CountVectorizer 사용하기**

- 띄어쓰기만을 기준으로 단어를 자름

- 길이가 2 이상인 문자에 대해서만 토큰으로 인식함


```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ['you know I want your love. because I love you.']
vector = CountVectorizer()

# 코퍼스로부터 각 단어의 빈도수를 기록
print('bag of words vector :', vector.fit_transform(corpus).toarray()) 

# 각 단어의 인덱스가 어떻게 부여되었는지를 출력
print('vocabulary :',vector.vocabulary_)
```

## 3-3. Document-Term Matrix, DTM

**[문서 단어 행렬(Document-Term Matrix, DTM)]**

- 다수의 문서에서 등장하는 각 단어들의 빈도를 행렬로 표현한 것.


- 각 문서에 대한 BoW를 하나의 행렬로 만든 것 (BoW에 대하여, 다수의 문서에 대한 행렬로 표현)

- 문서들을 서로 비교할 수 있도록 수치화할 수 있다는 점에서 의의를 가짐.

## 3-4. TF-IDF

**[TF-IDF(단어 빈도-역 문서 빈도, Term Frequency-Inverse Document Frequency)**]

* 단어의 빈도와 역 문서 빈도를 사용하여 DTM 내의 각 단어들마다 중요한 정도를 가중치로 주는 방법

* 모든 문서에서 자주 등장하는 단어는 중요도가 낮다고 판단하며, 특정 문서에서만 자주 등장하는 단어는 중요도가 높다고 판단

* TF-IDF는 주로 문서의 유사도를 구하는 작업, 검색 시스템에서 검색 결과의 중요도를 정하는 작업, 문서 내에서 특정 단어의 중요도를 구하는 작업 등에 사용

d(문서), t(단어), n(문서 총 개수)

**(1) tf(d,t) : 특정 문서 d에서의 특정 단어 t의 등장 횟수**

앞서 구한 DTM에서 각 단어들이 가진 값

**(2) df(t) : 특정 단어 t가 등장한 문서의 수**

즉, 특정 단어 t가 몇개의 문서에서 등장했는지(몇 번 등장했든지 간에, 등장한 문서 개수만 카운트)

**(3) idf(d, t) : df(t)에 반비례하는 수**

* TF-IDF는 특정 문서에서 자주 등장하는 단어는 그 문서 내에서 중요한 단어로 판단

* 문서2에서는 바나나를 한 번 언급했지만, 문서3에서는 바나나를 두 번 언급했기 때문에 문서3에서의 바나나를 더욱 중요한 단어라고 판단

```python
docs = [
  '먹고 싶은 사과',
  '먹고 싶은 바나나',
  '길고 노란 바나나 바나나',
  '저는 과일이 좋아요'
] 
```

```python
vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
vocab
```

```python
# 총 문서의 수
N = len(docs)
```

```python
from math import log

def tf(t, d):    # TF
  return d.count(t)

def idf(t):    # IDF
  df = 0
  for doc in docs:
    df += t in doc
  return log(N/(df+1))

def tfidf(t, d):    # TF-IDF
  return tf(t,d)* idf(t)
```

```python
# TF

import pandas as pd 
 
result = []

# 각 문서에 대해서 아래 연산을 반복
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
# IDF

result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))

idf_ = pd.DataFrame(result, index=vocab, columns=["IDF"])
idf_
```

```python
# TF-IDF

result = []
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

**유사도**

```python
#코사인 유사도

import torch

def cos_sim(A, B):
  return torch.dot(A, B)/(torch.norm(A)*torch.norm(B))
```
